```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Jakhongir0103/ML_Datasets/main/tashkent_houses_price.csv")
df.head()
```

```python
df.corrwith(df.price)
```

```python
plt.figure(figsize=(10,6))
sns.regplot( x=df[['size']], y=df[['price']])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2)
```

#### Simple Linear Regression

```python
x_simp_train = np.asanyarray(train[['size']])
y_simp_train = np.asanyarray(train[['price']])

x_simp_test = np.asanyarray(test[['size']])
y_simp_test = np.asanyarray(test[['price']])
```

```python
from sklearn import linear_model
LR_simple_model = linear_model.LinearRegression()

LR_simple_model.fit(x_simp_train, y_simp_train)
y_simp_predicted = LR_simple_model.predict(x_simp_test)
```

###### Evaluation

```python
from sklearn.metrics import mean_absolute_error
mea_simple = mean_absolute_error(y_simp_test, y_simp_predicted)
print(f'MAE: {mea_simple}')
```

### Multiple Linear Regression

```python
plt.figure(figsize=(10,6))
fig, ax = plt.subplots(1, 2, figsize=(15,6))
sns.regplot(ax=ax[0], x=df[['rooms']], y=df[['price']])
sns.regplot(ax=ax[1], x=df[['size']], y=df[['price']])
plt.show()
```

```python
x_mult_train = np.asanyarray(train[['rooms', 'size']])
y_mult_train = np.asanyarray(train[['price']])

x_mult_test = np.asanyarray(test[['rooms', 'size']])
y_mult_test = np.asanyarray(test[['price']])
```

```python
LR_model_multiple = linear_model.LinearRegression()

LR_model_multiple.fit(x_mult_train, y_mult_train)
y_multiple_predicted = LR_model_multiple.predict(x_mult_test)
```

###### Evaluation

```python
mea_mult = mean_absolute_error(y_mult_test, y_multiple_predicted)
print(f'MAE: {mea_mult}')
```
